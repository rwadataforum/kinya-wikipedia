# ANALYSIS - WIKIPEDIA 


this notebook compare kinyarwanda wikipedia with other african languages 

```python

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
from adjustText import adjust_text
```

```python
xFld_root = '/home/xmikelinux/__Projects/30_DataMgnt/30_WikipediaLLM/' 



xFld_charts = '/home/xmikelinux/__Projects/30_DataMgnt/30_WikipediaLLM/output/analysis/' 

FILE = xFld_root + 'data/wikipedia_languages_dataset.xlsx'
```

```python

# ── 1. Load & filter ──────────────────────────────────────────────────────────
df = pd.read_excel(FILE, sheet_name="Wikipedia Language Stats")

africa = df[df["Primary Continent(s)"].str.contains("Africa", na=False)].copy()
africa = africa[africa["Primary Continent(s)"].str.strip() == "Africa"]
africa = africa[~africa["Language"].isin(["French", "Arabic", "Portuguese"])]
africa["Continent Group"] = "Africa only"
africa = africa.dropna(subset=["Total Speakers (M)", "Articles"])
africa = africa[africa["Total Speakers (M)"] > 0]


africa.head()
```

```python


# ── 2. Seaborn theme ──────────────────────────────────────────────────────────
sns.set_theme(style="whitegrid", font="DejaVu Sans")
PALETTE = {"Africa only": "#2196A6"}

# ── 3. Figure & scatterplot ───────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 9))
fig.patch.set_facecolor("#F7F9FC")
ax.set_facecolor("#F7F9FC")

sns.scatterplot(
    data=africa,
    x="Total Speakers (M)",
    y="Articles",
    hue="Continent Group",
    palette=PALETTE,
    s=50,
    edgecolor="white",
    linewidth=0.6,
    alpha=0.9,
    ax=ax,
    zorder=4,
)

# ── 4. Log scale ──────────────────────────────────────────────────────────────
ax.set_xscale("log")
ax.set_yscale("log")

# ── 5. Labels with adjustText ─────────────────────────────────────────────────
label_colour = "#1A7A8A"

texts = []
for _, row in africa.iterrows():
    is_kinyarwanda = row["Language"] == "Kinyarwanda"
    t = ax.text(
        row["Total Speakers (M)"], row["Articles"],
        row["Language"],
        fontsize=9.5,
        color="#E02020" if is_kinyarwanda else label_colour,
        fontweight="bold",
        zorder=5,
    )
    texts.append(t)

if adjust_text is not None:
    adjust_text(
        texts, ax=ax,
        expand_points=(1.6, 2.0),
        expand_text=(1.3, 1.5),
        arrowprops=dict(arrowstyle="-", color="#BBBBBB", lw=0.5),
    )

# ── 6. Axis formatting ────────────────────────────────────────────────────────
def fmt_articles(x, _):
    if x >= 1_000_000: return f"{x/1_000_000:.1f}M"
    if x >= 1_000:     return f"{int(x/1000)}k"
    return str(int(x))

ax.yaxis.set_major_formatter(ticker.FuncFormatter(fmt_articles))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:g}"))
ax.yaxis.set_major_locator(ticker.LogLocator(base=10, numticks=10))
ax.xaxis.set_major_locator(ticker.LogLocator(base=10, numticks=8))
ax.yaxis.set_minor_locator(ticker.LogLocator(base=10, subs=np.arange(2, 10), numticks=50))
ax.xaxis.set_minor_locator(ticker.LogLocator(base=10, subs=np.arange(2, 10), numticks=30))
ax.yaxis.set_minor_formatter(ticker.NullFormatter())
ax.xaxis.set_minor_formatter(ticker.NullFormatter())
ax.grid(True, which="minor", linestyle=":", linewidth=0.3, color="#E0E0E0")
ax.tick_params(axis="both", which="major", labelsize=9, colors="#555555")

ax.set_xlabel("Total Speakers (millions)  [log scale]",
              fontsize=11, labelpad=10, color="#444444")
ax.set_ylabel("Wikipedia Articles  [log scale]",
              fontsize=11, labelpad=10, color="#444444")
ax.set_title(
    "African Language Editions of Wikipedia\n"
    "Article count vs. number of speakers  (log–log scale)",
    fontsize=14, fontweight="bold", color="#1A1A2E", pad=16,
)

# ── 7. Legend ─────────────────────────────────────────────────────────────────
ax.legend_.remove()

# ── 8. Footnote ───────────────────────────────────────────────────────────────
fig.text(
    0.5, -0.01,
    "Sources: en.wikipedia.org/wiki/List_of_Wikipedias (April 2026)  ·  "
    "Ethnologue via meta.wikimedia.org/wiki/List_of_Wikipedias_by_speakers_per_article",
    ha="center", fontsize=7, color="#AAAAAA",
)

plt.tight_layout()
plt.savefig(xFld_charts + "african_wikis_scatter.svg", dpi=180, bbox_inches="tight",
            facecolor=fig.get_facecolor())
print("Saved → african_wikis_scatter.svg")
plt.show()
```

## Closing the gap 

```python

# ── Compute ratio ─────────────────────────────────────────────────────────────
africa["ratio_speakers_art"] = (
    africa["Articles"] / africa["Total Speakers (M)"] * 1_000_000
)

# ── Sort descending; y-axis will be inverted so largest is at top ─────────────
by_articles = africa.sort_values("Articles", ascending=False).reset_index(drop=True)
by_ratio    = africa.sort_values("ratio_speakers_art", ascending=False).reset_index(drop=True)

# ── Colours ───────────────────────────────────────────────────────────────────
TEAL = "#2196A6"
RED  = "#E02020"
BG   = "#F7F9FC"

def bar_colours(df):
    return [RED if lang == "Kinyarwanda" else TEAL for lang in df["Language"]]

# ── Theme ─────────────────────────────────────────────────────────────────────
sns.set_theme(style="whitegrid", font="DejaVu Sans")

# ── Figure: two charts stacked vertically ────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(13, 22), facecolor=BG)
fig.patch.set_facecolor(BG)


def draw_hbar(ax, data, x_col, title, xlabel, fmt_fn):
    """Draw a styled horizontal bar chart on `ax`."""
    data = data.copy()
    data["_colour"] = bar_colours(data)

    sns.barplot(
        data=data, y="Language", x=x_col,
        hue="_colour", palette={TEAL: TEAL, RED: RED},
        legend=False, orient="h",
        ax=ax, edgecolor="white", linewidth=0.4,
    )
    ax.set_facecolor(BG)
    ax.set_title(title, fontsize=13, fontweight="bold", color="#1A1A2E", pad=12)
    ax.set_xlabel(xlabel, fontsize=10, color="#444444", labelpad=8)
    ax.set_ylabel("")

    # Y-axis tick labels — bold, Kinyarwanda in red
    ax.set_yticks(range(len(data)))
    ax.set_yticklabels(data["Language"], fontsize=10, fontweight="bold")
    for label in ax.get_yticklabels():
        label.set_color(RED if label.get_text() == "Kinyarwanda" else "#1A1A2E")

    ax.xaxis.set_major_formatter(ticker.FuncFormatter(fmt_fn))
    ax.tick_params(axis="x", labelsize=9, colors="#555555")
    ax.grid(axis="x", linestyle="--", linewidth=0.5, color="#CCCCCC")
    ax.set_axisbelow(True)

    # Value labels at end of each bar
    x_max = data[x_col].max()
    for patch in ax.patches:
        w = patch.get_width()
        if w > 0:
            ax.text(
                w + x_max * 0.005, patch.get_y() + patch.get_height() / 2,
                fmt_fn(w, None),
                ha="left", va="center", fontsize=8, color="#555555",
            )

    ax.set_xlim(right=x_max * 1.18)
    ax.invert_yaxis()
    for spine in ax.spines.values():
        spine.set_edgecolor("#DDDDDD")


# ── Top chart — by article count ─────────────────────────────────────────────
draw_hbar(
    ax1, by_articles, "Articles",
    title="Ranked by Number of Articles",
    xlabel="Number of Articles",
    fmt_fn=lambda x, _: f"{int(x/1000)}k" if x >= 1000 else str(int(x)),
)

# ── Bottom chart — by ratio ───────────────────────────────────────────────────
draw_hbar(
    ax2, by_ratio, "ratio_speakers_art",
    title="Ranked by Articles per Million Speakers",
    xlabel="Articles per Million Speakers",
    fmt_fn=lambda x, _: f"{int(x):,}",
)

# ── Suptitle & footnote ───────────────────────────────────────────────────────
plt.suptitle(
    "African Language Wikipedia Editions",
    fontsize=16, fontweight="bold", color="#1A1A2E", y=1.005,
)
fig.text(
    0.5, -0.005,
    "Sources: en.wikipedia.org/wiki/List_of_Wikipedias (April 2026)  ·  "
    "Ethnologue via meta.wikimedia.org/wiki/List_of_Wikipedias_by_speakers_per_article",
    ha="center", fontsize=7, color="#AAAAAA",
)

plt.tight_layout(h_pad=4)
plt.savefig(xFld_charts + "african_wikis_barcharts.svg", dpi=150, bbox_inches="tight",
            facecolor=fig.get_facecolor())
print("Saved → african_wikis_barcharts.svg")
plt.show()


```

```python
import pandas as pd 
import json 
import re 
import sqlite3
```

```python
# xdb 

xdb_file= '/home/xmikelinux/__ProjectsGitHub/kinya-wikipedia/output/analysis/analysis_db.sqlite'
```

```python

## fisrt create a datafram
```

```python
_KW_RE = re.compile(r"[^\s]+")

def _word_count(text: str) -> int:
    return len(_KW_RE.findall(text))

def _load_jsonl(path: str) -> pd.DataFrame:
    records = []
    with open(path, encoding="utf-8") as fh:
        for line in fh:
            line = line.strip()
            if not line:
                continue
            records.append(json.loads(line))
    df = pd.DataFrame(records)
    df["word_count"] = df["text"].apply(_word_count)
    df["char_count"] = df["text"].str.len()
    return df

xFile_jsonl = '/home/xmikelinux/__ProjectsGitHub/kinya-wikipedia/output/wikipedia/rw_0.jsonl'

q1 = _load_jsonl(xFile_jsonl) 

xcols_1 = ['id', 'title', 'url', 'word_count', 'char_count']
xcols_2 = ['article_id', 'title_kin', 'url_kin', 'word_count_kin', 'char_count_kin']

q1 = q1[xcols_1]
q1.columns = xcols_2

q1.head()
```

```python
# Connect to SQLite database (creates the file if it doesn't exist)
conn = sqlite3.connect(xdb_file)

# Save the DataFrame to SQLite
# if_exists options: 'fail', 'replace', 'append'
q1.to_sql('wikipedia_kin', conn, if_exists='replace', index=False)

# Close the connection
conn.close()

print(f"DataFrame saved to '{xdb_file}' in table 'q1_table'.")
```

```python
the subject_id (unique), then concatenate object_labels in a semicolone separated string and then add the most frequent object_label as xobject to it 
```

```python
## IN SQLITE 


CREATE TABLE wikidata_aggregate_kin AS
WITH global_counts AS (
    -- Count frequency of each object_label across entire dataset
    SELECT 
        object_label,
        COUNT(*) as global_cnt
    FROM main.wikidata_kin_triplets_p31
    GROUP BY object_label
),
subject_best_label AS (
    -- For each subject, find which of its labels has highest global frequency
    SELECT 
        s.subject_id,
        s.object_label as best_label,
        ROW_NUMBER() OVER (
            PARTITION BY s.subject_id 
            ORDER BY gc.global_cnt DESC, s.object_label ASC
        ) as rn
    FROM main.wikidata_kin_triplets_p31 s
    JOIN global_counts gc ON s.object_label = gc.object_label
)
SELECT 
    s.subject_id,
    GROUP_CONCAT(s.object_label, '; ') AS object_labels_concat,
    sbl.best_label AS xobject
FROM main.wikidata_kin_triplets_p31 s
JOIN subject_best_label sbl ON s.subject_id = sbl.subject_id AND sbl.rn = 1
GROUP BY s.subject_id;
```
